```python
# EMG data recorded at 200Hz
# IMU data recorded at 50Hz
# First need to bring the data to an equal number of samples
from scipy.ndimage import interpolation
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

data_emg = pd.read_csv('/home/amaya/emp/Realtime-Sign-Language-Translation-to-Speech-DNN/bird1/emg-1568110242.csv')
data_gyro = pd.read_csv('/home/amaya/emp/Realtime-Sign-Language-Translation-to-Speech-DNN/bird1/gyro-1568110242.csv')
data_acc = pd.read_csv('/home/amaya/emp/Realtime-Sign-Language-Translation-to-Speech-DNN/bird1/accelerometer-1568110242.csv')
emg = data_emg.loc[:,'emg1':]
gyro = data_gyro.loc[:,'x':]
acc = data_acc.loc[:,'x':]

i = len(data_gyro['timestamp'])
z = i/len(emg['emg1'])
output = []
for column in emg.columns:
    output.append(list(interpolation.zoom(emg[column],z)))

emgt = pd.DataFrame(np.transpose(output), columns = emg.columns)
```

```python
# bringing down the emg samples to the same number of samples as of gyro and acc
for column in emgt.columns:
    #ax = plt.gca()
    emg.plot(kind='line',y= column)
    emgt.plot(kind='line',y= column, color='red')
    plt.show()
```

```python
emgt.shape
```

```python
i_new = 50
z_new = i_new/len(emgt['emg1'])
emg_out = []
gyro_out = []
acc_out = []
for column in emgt.columns:
    emg_out.append(list(interpolation.zoom(emgt[column],z_new)))
for column in gyro.columns:
    gyro_out.append(list(interpolation.zoom(gyro[column],z_new)))
for column in acc.columns:
    acc_out.append(list(interpolation.zoom(acc[column],z_new)))    

emg_final = pd.DataFrame(np.transpose(emg_out),columns = emg.columns)
gyro_final = pd.DataFrame(np.transpose(gyro_out),columns = ['GXR','GYR','GZR'])
acc_final = pd.DataFrame(np.transpose(acc_out),columns = ['AXR','AYR','AZR'])

final = pd.concat([emg_final, acc_final, gyro_final],axis = 1)
```

```python
# 300+ timestamps vs 50
for column in emgt.columns:
    #ax = plt.gca()
    emgt.plot(kind='line',y= column)
    emg_final.plot(kind='line',y= column, color='red')
    plt.show()
```
